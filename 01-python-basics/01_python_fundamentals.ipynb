## Static vs Dynamic Typing
**Static Typing:** The type of a variable is checked at compile-time. Languages like Java or C++ require you to declare variable types explicitly (int x = 5;).

**Dynamic Typing (Python):** The type of a variable is checked at runtime. In Python, you don’t declare types; the interpreter figures it out when the program runs:

```python
x = 5       # integer
x = "hello" # now a string
```

```python
# C/C++
name = 'Aqib'
print(name)

a = 5
b = 6

print(a + b)
```

```python
# Dynamic Typing
a = 5
# Static Typing
int a = 5
```

## Static vs Dynamic Binding
**Static Binding:** The method or function call is resolved at compile-time. Example: function overloading in C++.

**Dynamic Binding (Python):** The method or function call is resolved at runtime. Python uses dynamic binding for methods:

```python
class Animal:
    def speak(self):
        print("Some sound")

class Dog(Animal):
    def speak(self):
        print("Bark")

pet = Dog()
pet.speak()  # resolved at runtime → "Bark"
```

```python
# Dynamic Binding
a = 5
print(a)
a = 'Aqib'
print(a)

# Static Binding
# int a = 5
```

# 4. Keywords & Identifiers

## Keywords
**Definition:** Keywords are reserved words in Python that have special meaning and cannot be used as variable names, function names, or identifiers.

```python
# Some Python keywords
if, else, while, for, def, class, return, import, True, False, None
```

## Identifiers
**Definition:** Identifiers are names given to variables, functions, classes, or objects in Python. They are created by the programmer.

**Rules:**

1. Must start with a letter or underscore (_)

2. Cannot start with a number

3. Cannot use keywords

4. Can contain letters, digits, and underscores

```python
# Identifiers
# You can't start with a digit
name1 = 'Aqib'
print(name1)
# You can use special chars -> _
_ = 'aqib'
print(_)
# identiers can not be keyword
```
